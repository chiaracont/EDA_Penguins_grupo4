```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

#Carga del dataset desde archivo local
df = pd.read_csv('../02_limpieza/penguins_limpio.csv')
```

```python
# ¿Cuántas especies hay?

df["Species"].value_counts()
```

```python
df['Species'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(4, 4),
    ylabel=''
)
plt.title('Proporción de Especies')
plt.show()
```

```python
# ¿Cómo varía la longitud del pico? (media, mediana, percentil en general)

df['Culmen Length (mm)'].describe()
```

```python
# ¿Cómo varía la profundidad del pico? (media, mediana, percentil en general)

df['Culmen Depth (mm)'].describe()
```

```python
# ¿Cómo varía la longitud de las aletas? (media, mediana, percentil en general)

df['Flipper Length (mm)'].describe()
```

```python
# ¿Cómo varía la masa corporal? (media, mediana, percentil en general)

df['Body Mass (g)'].describe()
```

```python
# ¿Cuántos pingüinos hay por sexo?

df['Sex'].value_counts()
```

```python
# ¿Cuántos pingüinos hay en cada isla?

df['Island'].value_counts()
```

```python
df['Island'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(4, 4),
    ylabel=''
)
plt.title('Pingüinos por islas')
plt.show()
```
